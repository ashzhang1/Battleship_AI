```python
import numpy as np
import random


def combine(s1,s2,s3,s4,s5):
    combined = np.zeros((10, 10), dtype=int)
    for x in range(0,10):
        for y in range(0,10):
            combined[x][y] = s1[x][y] + s2[x][y] + s3[x][y] + s4[x][y] + s5[x][y]
            if combined[x][y] > 1:
                return False
    return combined
    
def check_combo(s1,s2=np.zeros((10, 10)),s3=np.zeros((10, 10)),s4=np.zeros((10, 10)),s5=np.zeros((10, 10))):
    for x in range(0,10):
        for y in range(0,10):
            test = s1[x][y] + s2[x][y] + s3[x][y] + s4[x][y] + s5[x][y]
            if test > 1:
                return False
    return True

def combineCombinations(configsPerShip):
    all_combinations = []
    for s1 in configsPerShip[0]:
        for s2 in configsPerShip[1]:
            print("second one")
            if not check_combo(s1,s2):
                continue
            for s3 in configsPerShip[2]:
                if not check_combo(s1,s2,s3):
                    continue
                for s4 in configsPerShip[3]:
                    if not check_combo(s1,s2,s3,s4):
                        continue
                    for s5 in configsPerShip[4]:
                        combined = combine(s1,s2,s3,s4,s5)
                        if type(combined) != bool:
                            all_combinations.append(combined)

    return all_combinations      


def generateAllBoardStates():
    currentAllStates = []
    ships = [5, 3, 3, 2, 2]
    for ship in ships:
        currentAllStates.append(generateBoardStates(ship))
        print("length of ship ", ship, len(currentAllStates[-1]))
        
    all_combinations = combineCombinations(currentAllStates)

    return all_combinations




def get_ship_pos(x,y,ship_size,direction):
    ship_spots = []
    if direction == 0:
        for i in range(ship_size):
            ship_spots.append((x+i,y))
    elif direction == 1:
        for i in range(ship_size):
            ship_spots.append((x,y+i))
    return ship_spots

def pickStart():
    board = np.zeros((10, 10), dtype=int)
    ship_pos = []
    ships = [5, 3, 3, 2, 2]
    for ship in ships:
        board_possibilities, pos_possibilities = generateBoardStates(ship, board)
        board_index = random.choice(list(range(len(board_possibilities))))
        board = board_possibilities[board_index]
        pos = pos_possibilities[board_index]
        ship_pos.append(pos)

    return ship_pos

def generateBoardStates(shipSize, before_board):
    board_size = 10
    ship_sizes = [shipSize]
    board_states = []
    ship_spots = []
    
    def is_valid_placement(board, size, x, y, direction):
        if direction == 0:
            return x + size <= board_size and np.all(board[y, x:x+size] == 0)
        elif direction == 1:
            return y + size <= board_size and np.all(board[y:y+size, x] == 0)
        return False
    
    def place_ship(board, size, x, y, direction):
        if direction == 0:
            board[y, x:x+size] = 1
        elif direction == 1:
            board[y:y+size, x] = 1
    
    def remove_ship(board, size, x, y, direction):
        if direction == 0:
            board[y, x:x+size] = 0
        elif direction == 1:
            board[y:y+size, x] = 0
    
    def generate(board, remaining_ships):
        if not remaining_ships:
            #board_states.append(np.ravel(board.copy()))
            board_states.append(board.copy())
            return
        
        ship_size = remaining_ships[0]
        
        for x in [1,2,3,6,7,8]:
            for y in [1,2,3,6,7,8]:
                for direction in [0, 1]:
                    if is_valid_placement(board, ship_size, x, y, direction):
                        place_ship(board, ship_size, x, y, direction)
                        generate(board, remaining_ships[1:])
                        ship_spots.append(get_ship_pos(x,y,ship_size,direction))
                        remove_ship(board, ship_size, x, y, direction)
    
    generate(before_board, ship_sizes)
    
    return np.array(board_states), ship_spots

#generateAllBoardStates()[1].reshape((10, 10))
```
